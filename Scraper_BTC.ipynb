```python
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from time import sleep
from itertools import chain
DRIVER_PATH='./chromedriver'
options = Options()
options.headless = True
column_names = ["Date", "Text", "Number_of_likes", "Number_of_dislikes"]
```

```python
list = chain(range(110,190),range(750,1610),range(5215,5800))
```

```python
DF = pd.DataFrame(columns = column_names)
driver = webdriver.Chrome(executable_path = DRIVER_PATH)
path = "https://www.investing.com/crypto/bitcoin/chat/"+str(85)
driver.get(path)
sleep(30)
for i in list:
    dates =[]
    comments = []
    likes = []
    dislikes = []
    print(i)
    path = "https://www.investing.com/crypto/bitcoin/chat/"+str(i)
    driver.get(path)
    try:
        one_trust = driver.find_element_by_id("onetrust-accept-btn-handler").click()
    except:
        pass
    buttons=driver.find_elements_by_class_name('showMoreText')
    for button in buttons:
        try:
            button.click()
        except:
            pass
    X_comments = driver.find_elements_by_class_name("mainComment")
    for comment in X_comments:
        if comment.text.split("\n")[1] != "Just Now":
            try:
                dates.append(comment.text.split("\n")[1])
            except:
                dates.append(np.nan)
            try:
                comments.append(comment.text.split("\n")[2].split("Show more")[0])
            except:
                comments.append(np.nan)
            try:
                likes.append(comment.text.split("\n")[3].split(" ")[1])
            except:
                likes.append(np.nan)
            try:
                dislikes.append(comment.text.split("\n")[3].split(" ")[2])
            except:
                dislikes.append(np.nan)
        dictionary = {"Date": dates, "Text": comments, 
                    "Number_of_likes": likes, "Number_of_dislikes": dislikes}
    driver.close
    df = pd.DataFrame(dictionary)
    DF = DF.append(df, ignore_index=True)
```
